```python
invalids = []
for ticket in tickets:
    list_ticket = ticket.split(',')
    list_ticket = [int(i) for i in list_ticket]
    for ticket_component in list_ticket:
        is_valid = False
        for condition in list_conditions:
            (f1, c1), (f2, c2) = condition
            f1, c1, f2, c2 = int(f1), int(c1), int(f2), int(c2)
            #print('does ', ticket_component, ' fulfill these conditions', f1, c1, f2, c2)
            #print(((f1 <= ticket_component and ticket_component <= c1) or (f2 <= ticket_component and ticket_component <= c2)))
            is_valid = is_valid or ((f1 <= ticket_component and ticket_component <= c1) or (f2 <= ticket_component and ticket_component <= c2))
        if not is_valid:
            invalids.append(ticket_component)
```

```python
import numpy as np
```

```python
with open('input.txt') as f:
    ls = [line.strip() for line in f.readlines()]

# this is a better way to find the data we are interested in
ranges = [list(map(int, re.findall('\d+', x))) for x in ls[:20]]
your = np.array([int(x) for x in ls[22].split(',')], dtype=np.int64)
nearby = [list(map(int, re.findall('\d+', x))) for x in ls[25:]]
```

```python
valid = set()
# We can break this verification down into is this number inside of this set
# We are checking every number to see if it is in this valid set, so we can construct a valid set like this rather than making these hard rules

for t1, t2, t3, t4 in ranges:
    valid |= set(range(t1, t2 + 1))
    valid |= set(range(t3, t4 + 1))
# This is pretty neat, need to remember that you can do this with for loops
# You do (thing you want) for (big item) in (big generator) for (small item) in (big item) if statement
print(sum(n for l in nearby for n in l if n not in valid))
```

```python
# redefine the valids
# basically all(n in valid for n in l) will check to see that all of the values
# that are in l are in our list of all valid number set which was constructed before
# then if any of them aren't we get rid of the whole lot of them

valids = [l for l in nearby if all(n in valid for n in l)]
```

```python
from scipy.sparse import csr_matrix
from scipy.sparse.csgraph import maximum_bipartite_matching
```

```python
loc = [[all((t1 <= l[j] <= t2) or (t3 <= l[j] <= t4) for l in valids)
        for t1, t2, t3, t4 in ranges]
       for j in range(20)]
print(loc)
m = maximum_bipartite_matching(csr_matrix(loc))
print(your[m[:6]].prod())
```
