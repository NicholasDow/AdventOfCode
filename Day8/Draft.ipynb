```python
file = open('input.txt').readlines()
```

```python
visited = [False for _ in file]
```

```python
total_value = 0
```

```python
place = 0
```

```python
maxplace = len(file)
```

```python
while maxplace > place:
    print(file[place])
    if visited[place]:
        break
    visited[place] = True
    command = file[place][:3]
    value = int(file[place][4:])
    if command == 'acc':
        total_value += value
        place += 1
    elif command == 'nop':
        place += 1
    elif command == 'jmp':
        place += value
```

```python
temp_total = 0
temp_place = 0
temp_visited = visited
trying_something_new = False
```

```python
tried = [False for _ in file]
```

```python
while maxplace > place:
    print(file[place])
    if visited[place]:
        print("reset")
        total_value = temp_total
        place = temp_place
        visited = temp_visited
        trying_something_new = False
    visited[place] = True
    command = file[place][:3]
    value = int(file[place][4:])
    if command == 'acc':
        total_value += value
        place += 1
    elif command == 'nop':
        if not tried[place] and not trying_something_new:
            print("switching a nop")
            trying_something_new = True
            tried[place] = True
            temp_total = total_value
            temp_place = place
            temp_visited = visited.copy()
            place += value
        else:
            place += 1
    elif command == 'jmp':
        if not tried[place] and not trying_something_new:
            print("swithing a jmp")
            trying_something_new = True
            tried[place] = True
            temp_total = total_value
            temp_place = place
            temp_visited = visited.copy()
            place += 1
        else:
            place += value
print(maxplace, place)
```
