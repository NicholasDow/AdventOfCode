```python
def read_input(filename):
    with open(filename) as file:
        input = file.readlines()
    return [x.strip() for x in input]
```

```python
read_input('input.txt')
```

```python
def problem3_2():
    input = read_input('input.txt')
    slopes = [[1,1],[3,1],[5,1],[7,1],[1,2]]
    counts = []
    xlen = len(input[0])
    ylen = len(input)
    for slope in slopes:
        posx = 0
        posy = 0
        count = 0
        while posy < ylen - 1:
            posx += slope[0]
            if posx >= xlen:
                posx -= xlen
            posy += slope[1]
            if input[posy][posx] == '#':
                count += 1
        counts.append(count)
    print(counts)
    print(np.product(counts))
    
```
