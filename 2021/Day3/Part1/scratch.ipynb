```python
import numpy as np

lines = open('input.txt').read().split('\n')[:-1]
i = [[int(c) for c in list(x)] for x in lines]
i = np.array(i)
rows, cols = i.shape
oxy = i[:]
scb = i[:]
for col in range(cols):
    if oxy.shape[0] != 1:
        oxy_n = np.bincount(oxy[:,col])
        if oxy_n[0] == oxy_n[1]:
            oxy_n1 = 1
        else:
            oxy_n1 = oxy_n.argmax()
        oxy = oxy[oxy[:,col] == oxy_n1]

    if scb.shape[0] != 1:
        scb_n = np.bincount(scb[:,col])
        if scb_n[0] == scb_n[1]:
            scb_n1 = 0
        else:
            scb_n1 = scb_n.argmin()
        scb = scb[scb[:,col] == scb_n1]
        
        
```
