```python
text = open('input.txt', 'r')
```

```python
passwords = text.readlines()
```

```python
newpasswords = [x.strip('\n') for x in passwords]
```

```python
newpasswords
```

```python
import re
```

```python
li = []
for password in newpasswords:
    minimum = re.search('(\d+)[-]', password).group(1)
    maximum = re.search('[-](\d+)',password).group(1)
    letter = re.search('(\w)[:]', password).group(1)
    actualpassword = re.search('[:] (\w+)', password).group(1)
    
    li.append((int(minimum) <= actualpassword.count(letter)) and (int(maximum) >= actualpassword.count(letter)))
sum(li)
```
