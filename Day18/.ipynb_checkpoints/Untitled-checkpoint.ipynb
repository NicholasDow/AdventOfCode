```python
class backwards(int):
    def __mul__(self, b):
        return backwards(int(self) + b)
    # I found out later that you still actually need this for some reason, because you run sum????
    def __add__(self, b):
        return backwards(int(self) + b)
    def __sub__(self, b):
        return backwards(int(self) * b)
```

```python
def more_fuckery(line):
    line = re.sub(r"(\d+)", r"backwards(\1)", line)
    line = line.replace("*", "-")
    line = line.replace("+", "*")
    return eval(line, {}, {"backwards": backwards})
```
