```python
file = open('input.txt').readlines()
```

```python
file = [i.strip() for i in file]
```

```python
file
```

```python
direction = 'E'
```

```python
right = ['E','S','W','N']
left = ['E','N','W','S']
```

```python
X = 0
Y = 0
```

```python
for command in file:
    action = command[:1]
    degree = int(command[1:])
    if action == 'N':
        Y += degree
    elif action == 'S':
        Y -= degree
    elif action == 'E':
        X += degree
    elif action == 'W':
        X -= degree
    elif action == 'R':
        rot = degree/90
        direction = right[int((right.index(direction) + rot) % 4)]
    elif action == 'L':
        rot = degree/90
        direction = left[int((left.index(direction) + rot) % 4)]
    elif action == 'F':
        if direction == 'N':
            Y += degree
        elif direction == 'S':
            Y -= degree
        elif direction == 'E':
            X += degree
        elif direction == 'W':
            X -= degree
        
```
