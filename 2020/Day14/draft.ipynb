```python
with open('input.txt') as f:
    file = f.readlines()
    file = [i.strip().split(' = ') for i in file]
```

```python
file
```

```python
bitmask = ''
```

```python
# Oh man this is neat
def apply_mask(arg, mask):
    arg |= int(mask.replace('X', '0'), 2)
    arg &= int(mask.replace('X', '1'), 2)
    return arg     
```

```python
from collections import defaultdict
```

```python
mem = defaultdict(set)
```

```python
import re
```

```python
for command in file:
    if command[0] == 'mask':
        bitmask = command[1]
    else:
        mem[re.search('(\d+)',command[0]).group(0)] = apply_mask(int(command[1]), bitmask)
```

```python
sum(mem.values())
```

```python
mem = defaultdict(set)
```

```python
def bit_address(mask):
    if not mask:
        yield ''
        return
    for m in bit_address(mask[1:]):
        if mask[0] == '0':    
            yield 'X' + m
        elif mask[0] == '1':
            yield '1' + m
        elif mask[0] == 'X':
            yield '1' + m
            yield '0' + m
```

```python
for command in file:
    if command[0] == 'mask':
        bitmask = command[1]
    else:
        position = int(re.search('(\d+)',command[0]).group(0))
        for mask in bit_address(bitmask):
            mem[apply_mask(position, mask)] = int(command[1])
            
```

```python
sum(mem.values())
```
