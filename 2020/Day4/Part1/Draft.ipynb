```python
def read_input(filename):
    with open(filename) as f:
        input = f.read()
        return input
```

```python
file = read_input('input.txt')
```

```python
newfile = file.split('\n\n')
```

```python
newnewfile = [passport.replace('\n', ' ') for passport in newfile]
```

```python
newnewfile = [passport.split(' ') for passport in newnewfile]
```

```python
def list_to_dict(li):
    print(li)
    return dict(e.split(':') for e in li)
```

```python
def remove_spaces(li):
    while("" in li) : 
        li.remove("") 
    return li
```

```python
newnewfile = [remove_spaces(passport) for passport in newnewfile]
```

```python
dictfile = [list_to_dict(passport) for passport in newnewfile]
```

```python
import re
```

```python
def check_height(height):
    if height[-2:] == 'cm':
        height = int(height[:-2])
        return (height >= 150) and (height <= 193)
    if height[-2:] == 'in':
        height = int(height[:-2])
        return (height >= 59) and (height <= 76)
    else:
        return False
```

```python
def check_hair(hair):
    return hair[0] == '#' and len(hair) == 7 and re.search('(?:[0-9a-f]+)$', hair[1:])
```

```python
def check_id(x):
    return len(x) == 9 and x.isdigit()
```

```python
def check_eye(x):
    return x in ['amb','blu','brn','gry','grn','hzl','oth']
```

```python
count = 0
for passport in dictfile:
    if set(['byr','iyr','eyr','hgt','hcl','ecl','pid']).issubset(set(passport.keys())):
        minage = 1920 <= int(passport['byr'])
        maxage = 2002 >= int(passport['byr'])
        minissue = 2010 <= int(passport['iyr'])
        maxissue = 2020 >= int(passport['iyr'])
        exprcheck = (len(passport['eyr']) == 4) and (int(passport['eyr']) >= 2020) and (int(passport['eyr']) <= 2030)
        fourdigits = (len(passport['byr']) == 4) and (len(passport['iyr']) == 4)
        hgtcheck = check_height(passport['hgt'])
        eyecheck = check_eye(passport['ecl'])
        haircheck = check_hair(passport['hcl'])
        idcheck = check_id(passport['pid'])
        if minage and maxage and minissue and maxissue and exprcheck and hgtcheck and eyecheck and idcheck and haircheck :
            count += 1
print(count)
```
