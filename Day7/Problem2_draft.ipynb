```python
file = open('input.txt','r')
file = file.readlines()
file = [rule.replace('bags','bag') for rule in file]
newfile = [rule.split('bag') for rule in file]
```

```python
li = len(newfile)*[None]
```

```python
for i, rule in enumerate(newfile):
    if '.\n' in rule:
        rule.remove('.\n')
    for j,s in enumerate(rule):
        s = s.replace('contain','')
        s = s.replace(',','')
        s = s.replace('.','')
        rule[j] = s.strip()
    # HAHAHA you cannot do this, you are modifying the file as you go, this will totally fuck up the ordering
    # newfile[i] = rule
    li[i] = rule
```

```python
li
```

```python
index = [first[0] for first in li]
```

```python
def number_of_bags(color):
    n = 1
    for i in li[index.index(color)][1:]:
        if i == 'no other':
            return 1
        else:
            n += int(i[0]) * number_of_bags(i[2:])
    return n
```
