# Part 1 can be solved with sets.

```python
with open('input.txt') as f:
    file = f.read()

# split into groups on the new line
file =  file.split('\n\n')
# create a set out of all the existing elements
solution = sum([len(set(f.replace('\n', ''))) for f in file])

```
